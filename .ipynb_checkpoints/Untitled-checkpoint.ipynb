```python
import dash
import dash_bootstrap_components as dbc
import dash_html_components as html
from pandas import DataFrame as df
import pandas as pd
from datetime import date, timedelta
yesterdays_date = (date.today() - timedelta(days=1)).strftime('%-m/%d/%y')
date_2_days_ago = (date.today() - timedelta(days=2)).strftime('%-m/%d/%y')
date_3_days_ago = (date.today() - timedelta(days=3)).strftime('%-m/%d/%y')
date_4_days_ago = (date.today() - timedelta(days=4)).strftime('%-m/%d/%y')
date_5_days_ago = (date.today() - timedelta(days=5)).strftime('%-m/%d/%y')
date_6_days_ago = (date.today() - timedelta(days=6)).strftime('%-m/%d/%y')
date_7_days_ago = (date.today() - timedelta(days=7)).strftime('%-m/%d/%y')
date_8_days_ago = (date.today() - timedelta(days=8)).strftime('%-m/%d/%y')

confirmed_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
deaths_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'

confirmed = pd.read_csv(confirmed_url)
deaths = pd.read_csv(deaths_url)
confirmed.iloc[0]['Country/Region'] #Test

def new_cases_columns(frame): #Makes 7 columns of new cases
    new_yest = [confirmed.iloc[i][yesterdays_date] - confirmed.iloc[i][date_2_days_ago] for i in range(len(frame))]
    new_2_ago = [confirmed.iloc[i][date_2_days_ago] - confirmed.iloc[i][date_3_days_ago] for i in range(len(frame))]
    new_3_ago = [confirmed.iloc[i][date_3_days_ago] - confirmed.iloc[i][date_4_days_ago] for i in range(len(frame))]
    new_4_ago = [confirmed.iloc[i][date_4_days_ago] - confirmed.iloc[i][date_5_days_ago] for i in range(len(frame))]
    new_5_ago = [confirmed.iloc[i][date_5_days_ago] - confirmed.iloc[i][date_6_days_ago] for i in range(len(frame))]
    new_6_ago = [confirmed.iloc[i][date_6_days_ago] - confirmed.iloc[i][date_7_days_ago] for i in range(len(frame))]
    new_7_ago = [confirmed.iloc[i][date_7_days_ago] - confirmed.iloc[i][date_8_days_ago] for i in range(len(frame))]
    confirmed['new_yest'] = df(new_yest)
    confirmed['new_2_ago'] = df(new_2_ago)
    confirmed['new_3_ago'] = df(new_3_ago)
    confirmed['new_4_ago'] = df(new_4_ago)
    confirmed['new_5_ago'] = df(new_5_ago)
    confirmed['new_6_ago'] = df(new_6_ago)
    confirmed['new_7_ago'] = df(new_7_ago)

new_cases_columns(confirmed)    

x = confirmed[confirmed["Country/Region"] == "Italy"].iloc[0]['new_yest']
print(x)

def change_past_week(frame):
    return frame.iloc[0]['new_yest']
new_cases_italy = change_past_week(confirmed[confirmed['Country/Region'] == 'Italy'])
print(new_cases_italy)
```
